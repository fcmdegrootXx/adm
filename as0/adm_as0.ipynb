Assignment 0 - Distinct_values

- N - number of buckets
- n - number of coins "thrown" into the buckets

- Probability of  GIVEN coin NOT hitting the bucket = ((N-1) / N)
- Probability of NONE coins hitting the bucket  = ((N-1) / N)^n 
- Approximation of small e > (1-q)^1/q = 1/e (Chapter 1.3.5) 
- In our case (1-1/N) = q. Therefore ((N - 1) / N)^n ≈ (1 - 1/N)^N ≈ 1/e > For each bucket the probability of not being hit 1/e
- For all buckets N * e^(-n/N)



```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define a range of N and n values to simulate
N_values = range(1, 401)  
n_values = range(1, 101)  

# Initialize an empty list to store results
results = []

# Iterate through different values of N and n
for N in N_values:
    for n in n_values:
        expected_distinct_values = N * math.exp(-n/N)
        results.append((N, n, expected_distinct_values))

# Separate the results for plotting
N_values = [result[0] for result in results]
n_values = [result[1] for result in results]
expected_values = [result[2] for result in results]

# Create a scatter plot to visualize the results
plt.figure(figsize=(8, 5))
plt.scatter(N_values, n_values, c=expected_values, cmap='viridis')
plt.colorbar(label='Expected Distinct Values')
plt.xlabel('N (Number of Buckets)')
plt.ylabel('n (Number of Coins Thrown)')
plt.title('Expected Distinct Values')
plt.grid(True)
plt.show()
```
